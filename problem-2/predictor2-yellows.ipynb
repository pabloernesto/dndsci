```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
data = pd.read_csv("sacrifices-clean.csv")
```

```python
non_yellows = data.query('color != "yellow"')

X = non_yellows.loc[:, "thmm"].values.reshape(-1, 1)
Y = non_yellows.loc[:, "mana"].values.reshape(-1, 1)

linear_regressor = LinearRegression()
linear_regressor.fit(X, Y)
```

```python
def predictor(color, thmm):
    if color == "yellow":
        return 19.5
    else:
        return linear_regressor.predict(np.array([thmm], ndmin=2))[0][0]
```

```python
data["pred"] = data.apply(lambda r: predictor(r["color"], r["thmm"]), axis=1)
data
```

```python
r2_score(data["mana"], data["pred"])
```
