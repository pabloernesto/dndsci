```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
data = pd.read_csv("sacrifices-clean.csv")
```

```python
blues = data.query('color == "blue"')

X = blues.loc[:, "thmm"].values.reshape(-1, 1)
Y = blues.loc[:, "mana"].values.reshape(-1, 1)

linear_regressor = LinearRegression()
linear_regressor.fit(X, Y)
```

```python
greens_reds = data.query('color in ["green", "red"]')
greens_reds["mana"].mean()
```

```python
def predictor(color, thmm):
    if color == "yellow":
        return 19.5
    elif color == "blue":
        return linear_regressor.predict(np.array([thmm], ndmin=2))[0][0]
    else:
        return 22.9
```

```python
data["pred"] = data.apply(lambda r: predictor(r["color"], r["thmm"]), axis=1)
data
```

```python
r2_score(data["mana"], data["pred"])
```

```python
print(linear_regressor.coef_, linear_regressor.intercept_)
```
